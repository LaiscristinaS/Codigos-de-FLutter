```python
import sqlite3
from sqlite3 import Error

# Função para executar comandos de alteração no banco
def executar_comando(conexao, sql):
    cursor = conexao.cursor()
    try:
        cursor.execute(sql)
        conexao.commit()
        print("Operação realizada com sucesso.")
    except Error as erro:
        print(f"Falha ao executar o comando: {erro}")

# Função para consultas
def consultar_dados(conexao, sql):
    cursor = conexao.cursor()
    try:
        cursor.execute(sql)
        resultado = cursor.fetchall()
        return resultado
    except Error as erro:
        print(f"Falha na consulta: {erro}")

# Abrindo conexão com o banco
conexao = sqlite3.connect("monitoria_ifmg.sqlite")
print("Conexão estabelecida com o banco de dados.")
```

```python
query = """
CREATE TABLE usuario (
    ra INTEGER PRIMARY KEY,
    nome VARCHAR(100) NOT NULL,
    email VARCHAR(100) NOT NULL UNIQUE,
    senha VARCHAR(255) NOT NULL,
    tipo TEXT NOT NULL
);
"""

execute_query(connection, query)

query = """
CREATE TABLE aluno (
    ra INT(20) PRIMARY KEY,
    turma VARCHAR(20) NOT NULL,
    FOREIGN KEY (ra) REFERENCES usuario(ra)
);
"""

execute_query(connection, query)

query = """
CREATE TABLE monitor (
    ra INT(20) PRIMARY KEY,
    FOREIGN KEY (ra) REFERENCES aluno(ra)
);
"""

execute_query(connection, query)

query = """
CREATE TABLE auditoria_admin (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    ra_admin_origem INT(20) NOT NULL,
    ra_usuario_promovido INT(20) NOT NULL,
    data_promocao DATETIME NOT NULL,
    FOREIGN KEY (ra_admin_origem) REFERENCES usuario(ra),
    FOREIGN KEY (ra_usuario_promovido) REFERENCES usuario(ra)
);
"""

execute_query(connection, query)

query = """
CREATE TABLE comunicado (
    id_comunicado INTEGER PRIMARY KEY AUTOINCREMENT,
    titulo VARCHAR(150) NOT NULL,
    conteudo TEXT NOT NULL,
    data_publicacao DATETIME NOT NULL,
    ra_admin INT(20) NOT NULL,
    FOREIGN KEY (ra_admin) REFERENCES usuario(ra)
);
"""

execute_query(connection, query)

query = """
CREATE TABLE monitoria (
    id_monitoria INTEGER PRIMARY KEY AUTOINCREMENT,
    disciplina VARCHAR(100) NOT NULL,
    local VARCHAR(100) NOT NULL,
    horario VARCHAR(50) NOT NULL,
    ra_monitor INT(20) NOT NULL,
    FOREIGN KEY (ra_monitor) REFERENCES monitor(ra)
);
"""

execute_query(connection, query)

query = """
CREATE TABLE aula_monitoria (
    id_aula INTEGER PRIMARY KEY AUTOINCREMENT,
    data_aula DATE NOT NULL,
    id_monitoria INT NOT NULL,
    FOREIGN KEY (id_monitoria) REFERENCES monitoria(id_monitoria)
);
"""

execute_query(connection, query)

query = """
CREATE TABLE presenca (
    id_presenca INTEGER PRIMARY KEY AUTOINCREMENT,
    id_aula INT NOT NULL,
    ra_aluno INT(20) NOT NULL,
    FOREIGN KEY (id_aula) REFERENCES aula_monitoria(id_aula),
    FOREIGN KEY (ra_aluno) REFERENCES aluno(ra),
    UNIQUE (id_aula, ra_aluno)
);
"""

execute_query(connection, query)

query = """
CREATE TABLE blog_monitoria (
    id_blog INTEGER PRIMARY KEY AUTOINCREMENT,
    id_monitoria INT NOT NULL,
    FOREIGN KEY (id_monitoria) REFERENCES monitoria(id_monitoria)
);
"""

execute_query(connection, query)

query = """
CREATE TABLE postagem_discussao (
    id_postagem INTEGER PRIMARY KEY AUTOINCREMENT,
    id_blog INT NOT NULL,
    ra_usuario INT(20) NOT NULL,
    nome_usuario VARCHAR(100) NOT NULL,
    conteudo TEXT NOT NULL,
    data_postagem DATETIME NOT NULL,
    FOREIGN KEY (id_blog) REFERENCES blog_monitoria(id_blog),
    FOREIGN KEY (ra_usuario) REFERENCES usuario(ra)
);
"""

execute_query(connection, query)

query = """
CREATE TABLE enquete (
    id_enquete INTEGER PRIMARY KEY AUTOINCREMENT,
    id_blog INT NOT NULL,
    titulo VARCHAR(150) NOT NULL,
    data_criacao DATETIME NOT NULL,
    FOREIGN KEY (id_blog) REFERENCES blog_monitoria(id_blog)
);
"""

execute_query(connection, query)

query = """
CREATE TABLE opcao_enquete (
    id_opcao INTEGER PRIMARY KEY AUTOINCREMENT,
    id_enquete INT NOT NULL,
    descricao VARCHAR(100) NOT NULL,
    FOREIGN KEY (id_enquete) REFERENCES enquete(id_enquete)
);
"""

execute_query(connection, query)

query = """
CREATE TABLE resposta_enquete (
    id_resposta INTEGER PRIMARY KEY AUTOINCREMENT,
    id_opcao INT NOT NULL,
    ra_usuario INT(20) NOT NULL,
    FOREIGN KEY (id_opcao) REFERENCES opcao_enquete(id_opcao),
    FOREIGN KEY (ra_usuario) REFERENCES usuario(ra),
    UNIQUE (id_opcao, ra_usuario)
);
"""

execute_query(connection, query)

query = """
CREATE TABLE notificacao (
    id_notificacao INTEGER PRIMARY KEY AUTOINCREMENT,
    id_blog INT NOT NULL,
    titulo VARCHAR(150) NOT NULL,
    conteudo TEXT NOT NULL,
    data_publicacao DATETIME NOT NULL,
    FOREIGN KEY (id_blog) REFERENCES blog_monitoria(id_blog)
);
"""

execute_query(connection, query)

query = """
CREATE TABLE notificacao_usuario (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    id_notificacao INT NOT NULL,
    ra_usuario INT(20) NOT NULL,
    visualizada BOOLEAN DEFAULT FALSE,
    FOREIGN KEY (id_notificacao) REFERENCES notificacao(id_notificacao),
    FOREIGN KEY (ra_usuario) REFERENCES usuario(ra)
);
"""

execute_query(connection, query)
```

## Atividade 05

Crie uma function relevante para sua aplicação em PI3.
Descreva o valor esperado como retorno ao executar a function e, ao justificar o contexto, explique por que é importante ter essa lógica encapsulada para uso recorrente no seu PI3.

* **Justificativa para que a operação seja uma Function (Contexto de PI3):** A verificação da frequência dos alunos é uma operação recorrente no sistema de monitoria, então, é importante guardar essa funcionalidade para reutilizá-la em relatórios e consultas do sistema.

* **Descrição da operação realizada e valor retornado pela Function:** A function recebe a quantidade de presenças do aluno e retorna uma classificação de frequência (Baixa, Média ou Alta).

```python
def avaliar_frequencia(qtd_presencas):

    if qtd_presencas <= 1:
        return "Frequência Baixa"

    elif qtd_presencas <= 3:
        return "Frequência Moderada"

    return "Frequência Alta"
```

```python
sql = """
SELECT
    u.nome AS Aluno,
    COUNT(p.id_presenca) AS Quantidade_Presencas
FROM usuario u
INNER JOIN aluno a
    ON u.ra = a.ra
INNER JOIN presenca p
    ON a.ra = p.ra_aluno
GROUP BY u.nome;
"""

dados = consultar_dados(conexao, sql)

for registro in dados:

    aluno = registro[0]
    presencas = registro[1]

    situacao = avaliar_frequencia(presencas)

    print(aluno, "-", presencas, "-", situacao)
```
